```python
with open('rosalind_lgis.txt', 'r') as f:
    f = f.readlines()
    n = int(f[0])
    π = list(map(int, f[1].split()))
```

```python
def lgis(π):
    incr = []
    decr = []
    for (i,x) in enumerate(π):
        a = max([incr[j] for (j, y) in enumerate(π[:i]) if y < x] or [[]], key=len) + [x]
        incr.append(a)
        
    for (i,x) in enumerate(π):
        b = max([decr[j] for (j, y) in enumerate(π[:i]) if y > x] or [[]], key=len) + [x]
        decr.append(b)
        
    print(*max(incr, key=len))
    print(*max(decr, key=len))
```

```python
lgis(π)
```
